```python
exponential_Func = lambda  x: np.exp(x)

# Sigmoid Function - Also called as squashing function
def sigmoid(input_x):
    return(1/(1+np.exp(-input_x)))

# Hard Sigmoid Function
def hard_sigmoid(input_x):
    return(np.max([0,np.min([1,(1+input_x)/2])]))
# hard sigmoid offers lesser computation cost when implemented bothe in a specalised hardware or software form
```

```python
# Values can be modified to see the extent of values
# By and large the values are similar, at lower computation cost
for i in range(0,20):
    print("Sigmoid for {}  - Sigmoid Val : {} and Hard_Sigmoid Val : {}".format(i,sigmoid(i),hard_sigmoid(i)))
hard_sigmoid(0)
```

```python
"""
Tanh function
Hyperbolic Tangent Function is a smoother zero-centred function
This is preferred since this gives better training performance for multi-layer neural networks
Tanh doesn't solve the vanishing gradient problem suffered by Sigmoid Function
The main advantage of tanh function is the zero centered output thereby aiding the back-propogation process
"""
tanh = lambda x : np.tanh(x)

in_array = np.linspace(-np.pi,np.pi,12)
out_array = tanh(in_array)

plt.plot(in_array,out_array)
```

```python
# Softmax - Used in finding the final class
def softmax(x):
    "Compute Softmax values for each sets of scores in input_x"
    e_x = np.exp(x - np.max(x)) 
    return e_x / e_x.sum(axis=0) 
```
